# El Algoritmo de Viterbi:

Imagina que estás observando a alguien actuar (las observaciones), pero no sabes qué está pensando (los estados ocultos). Sabes que sus pensamientos influyen en sus acciones y que un pensamiento suele llevar a otro. El Algoritmo de Viterbi es el detective que usa las matemáticas para encontrar la secuencia de pensamientos más probable que explica las acciones que viste.

Pertenece a la clase de algoritmos de Programación Dinámica. En lugar de calcular todas las combinaciones posibles de estados (lo cual sería computacionalmente imposible en secuencias largas), Viterbi desglosa el problema:

- Calcula la probabilidad máxima de llegar a cada estado en el tiempo $t$.
- Almacena el "camino" que lo llevó ahí.
- Al final, simplemente sigue el rastro hacia atrás (backtracking) para reconstruir la historia completa.

Los 3 Pilares del Modelo (HMM)

Para que el código funcione, necesitamos tres ingredientes:
- Probabilidades Iniciales ($\pi$): ¿Dónde es más probable que empiece todo?
- Matriz de Transición: La probabilidad de pasar del estado $A$ al $B$.
- Matriz de Emisión: La probabilidad de que, estando en el estado oculto $A$, observemos el evento $1$.

#Objetivo de la implementación

El objetivo principal de este código es resolver el problema de decodificación en un Modelo Oculto de Markov.Específicamente, busca:

- Optimización: Encontrar la secuencia de estados ocultos ($S_1, S_2, ..., S_n$) que maximice la probabilidad conjunta con la secuencia de observaciones dadas ($O_1, O_2, ..., O_n$).
- Eficiencia: Reducir la complejidad computacional de un crecimiento exponencial $O(N^T)$ a uno lineal respecto al tiempo $O(T \times N^2)$, donde $N$ es el número de estados y $T$ la longitud de la secuencia.

- Inferencia: Determinar la "explicación más plausible" de un fenómeno observado donde los factores causales no son directamente visibles.

```python
import numpy as np

#definimos el modelo oculto de Markov (HMM)
estados = ['A', 'B']
observaciones = ['1', '2', '3']

#probabilidades iniciales
pi={'A':0.6, 'B':0.4}

#Matriz de transicion
transiciones = {
    'A': {'A': 0.7, 'B': 0.3},
    'B': {'A': 0.4, 'B': 0.6}
}

#Matriz de emision
emisiones={
    'A': {'1': 0.1, '2': 0.4, '3': 0.5},
    'B': {'1': 0.7, '2': 0.2, '3': 0.1}
}

#Secuencia de observaciones
observaciones_seq = ['1','2','3']
```

```python
#Funcion para el algoritmo de Viterbi
def viterbi (obs,estados,pi,transiciones,emisiones):
  T = len(obs) #Longitud de observaciones
  N = len(estados) # Numero de estados

  #Inicializacion de las matrices V y path
  V = np.zeros((N,T))
  path = np.zeros((N,T),dtype=int)

  #Inixializacion de la primera columna de V
  for s in range(N):
    V[s][0]=pi[estados[s]] * emisiones[estados[s]][obs[0]]

  #calculo de las demas columnas de V y actualizacion de path
  for t in range(1,T):
    for s in range(N):
      max_prob = -1
      max_state = None
      for s_prev in range(N):
        prob = V[s_prev][t-1] * transiciones[estados[s_prev]][estados[s]]
        if prob > max_prob:
          max_prob = prob
          max_state = s_prev
      V[s][t] = max_prob
      path[s][t] = max_state

  #encontrar el estado mas probable en el ultimo paso
  max_prob = -1
  last_state = None
  for s in range(N):
    if V[s][T-1] > max_prob:
      max_prob = V[s][T-1]
      last_state = s
  #reconstruir la secuencia de estados mas probables
  stat_seq = [estados[last_state]]
  current_state = last_state
  for t in range(T-1,0,-1):
    current_state = path[current_state][t]
    stat_seq.insert(0,estados[current_state])

  return stat_seq

#Ejecutar el algoritmo de Viterbi
resultado = viterbi(observaciones_seq,estados,pi,transiciones,emisiones)
print("Secuencia de estados mas probable: ", resultado)
```
