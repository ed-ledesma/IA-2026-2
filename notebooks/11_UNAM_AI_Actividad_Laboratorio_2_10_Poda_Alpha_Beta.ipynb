```python
import math
import time

inicio = time.time()
def imprimir_tablero(tablero):
    for fila in tablero:
        print(" ".join(fila))
    print()

def es_tablero_lleno(tablero):
    for fila in tablero:
        if " " in fila:
            return False
    return True

def evaluar_estado(tablero):
    # Evaluación simple para el tic-tac-toe
    ganador = None
    # Filas
    for fila in tablero:
        if fila.count("X") == 3:
            ganador = "X"
        elif fila.count("O") == 3:
            ganador = "O"
    # Columnas
    for col in range(3):
        if tablero[0][col] == tablero[1][col] == tablero[2][col] and tablero[0][col] != " ":
            ganador = tablero[0][col]
    # Diagonales
    if tablero[0][0] == tablero[1][1] == tablero[2][2] and tablero[0][0] != " ":
        ganador = tablero[0][0]
    if tablero[0][2] == tablero[1][1] == tablero[2][0] and tablero[0][2] != " ":
        ganador = tablero[0][2]

    if ganador == "X":
        return 1
    elif ganador == "O":
        return -1
    elif es_tablero_lleno(tablero):
        return 0
    else:
        return None

def alphabeta(tablero, profundidad, alfa, beta, es_maximizando):
    estado_actual = evaluar_estado(tablero)

    if estado_actual is not None:
        return estado_actual

    if es_maximizando:
        mejor_valor = -math.inf
        for i in range(3):
            for j in range(3):
                if tablero[i][j] == " ":
                    tablero[i][j] = "X"
                    valor = alphabeta(tablero, profundidad + 1, alfa, beta, False)
                    tablero[i][j] = " "
                    mejor_valor = max(mejor_valor, valor)
                    alfa = max(alfa, mejor_valor)
                    if beta <= alfa:
                        break
        return mejor_valor
    else:
        mejor_valor = math.inf
        for i in range(3):
            for j in range(3):
                if tablero[i][j] == " ":
                    tablero[i][j] = "O"
                    valor = alphabeta(tablero, profundidad + 1, alfa, beta, True)
                    tablero[i][j] = " "
                    mejor_valor = min(mejor_valor, valor)
                    beta = min(beta, mejor_valor)
                    if beta <= alfa:
                        break
        return mejor_valor

def mejor_movimiento(tablero):
    mejor_valor = -math.inf
    mejor_mov = None
    alfa = -math.inf
    beta = math.inf

    for i in range(3):
        for j in range(3):
            if tablero[i][j] == " ":
                tablero[i][j] = "X"
                valor = alphabeta(tablero, 0, alfa, beta, False)
                tablero[i][j] = " "

                if valor > mejor_valor:
                    mejor_valor = valor
                    mejor_mov = (i, j)
                alfa = max(alfa, mejor_valor)

    return mejor_mov

# Ejemplo de uso con tic-tac-toe
tablero_inicial = [
    [" ", " ", " "],
    [" ", " ", " "],
    [" ", " ", " "]
]

turno = "X"

while evaluar_estado(tablero_inicial) is None and not es_tablero_lleno(tablero_inicial):
    imprimir_tablero(tablero_inicial)

    if turno == "X":
        fila, col = map(int, input("Turno de X. Ingrese fila y columna (separadas por espacio): ").split())
        if tablero_inicial[fila][col] == " ":
            tablero_inicial[fila][col] = "X"
        else:
            print("Casilla ocupada. Intenta de nuevo.")
            continue
    else:
        print("Turno de O (IA)")
        movimiento = mejor_movimiento(tablero_inicial)
        fila, col = movimiento
        tablero_inicial[fila][col] = "O"

    turno = "O" if turno == "X" else "X"

imprimir_tablero(tablero_inicial)
resultado = evaluar_estado(tablero_inicial)

if resultado == 1:
    print("¡X gana!")
elif resultado == -1:
    print("¡O gana!")
else:
    print("¡Empate!")
fin = time.time()
print("Tiempo:", fin - inicio)
```

```python
# IMPORTACIÓN DE LIBRERÍAS
import math   # Se utiliza para trabajar con infinito (math.inf)
import time   # Se utiliza para medir el tiempo de ejecución del programa


# VARIABLES GLOBALES DE CONFIGURACIÓN

nodos_explorados = 0
# Contador de nodos explorados por el algoritmo.
# Sirve para medir cuántos estados del juego analiza el algoritmo.

USAR_PODA = False
# Variable booleana que indica si se utilizará o no la poda Alpha-Beta.
# False -> ejecuta Minimax normal
# True  -> ejecuta Minimax con poda Alpha-Beta

PROFUNDIDAD_MAX = 3
# Profundidad máxima de búsqueda del algoritmo.
# Limita cuántos niveles del árbol de juego se explorarán.


# FUNCIÓN: imprimir_tablero
# Muestra el estado actual del tablero en consola
def imprimir_tablero(tablero):
    for fila in tablero:
        print(" ".join(fila))
    print()


# FUNCIÓN: es_tablero_lleno
# Verifica si el tablero ya no tiene espacios vacíos
# Retorna True si el tablero está lleno
def es_tablero_lleno(tablero):
    for fila in tablero:
        if " " in fila:
            return False
    return True


# FUNCIÓN: evaluar_estado
# Evalúa el estado actual del juego.
#
# Retorna:
#  10  -> gana X
# -10  -> gana O
#   0  -> nadie ha ganado todavía
def evaluar_estado(tablero):

    # Revisar filas
    for fila in tablero:
        if fila.count("X") == 3:
            return 10
        if fila.count("O") == 3:
            return -10

    # Revisar columnas
    for col in range(3):
        if tablero[0][col] == tablero[1][col] == tablero[2][col] and tablero[0][col] != " ":
            if tablero[0][col] == "X":
                return 10
            else:
                return -10

    # Revisar diagonal principal
    if tablero[0][0] == tablero[1][1] == tablero[2][2] and tablero[0][0] != " ":
        return 10 if tablero[0][0] == "X" else -10

    # Revisar diagonal secundaria
    if tablero[0][2] == tablero[1][1] == tablero[2][0] and tablero[0][2] != " ":
        return 10 if tablero[0][2] == "X" else -10

    return 0


# FUNCIÓN: alphabeta
#
# Implementa el algoritmo Minimax con opción de poda Alpha-Beta.
#
# Parámetros:
# tablero -> estado actual del juego
# profundidad -> nivel actual del árbol de búsqueda
# alfa -> mejor valor encontrado para el jugador MAX
# beta -> mejor valor encontrado para el jugador MIN
# es_max -> indica si el turno es del jugador MAX
#
# El algoritmo explora posibles jugadas futuras para elegir
# la mejor decisión posible.
def alphabeta(tablero, profundidad, alfa, beta, es_max):

    global nodos_explorados
    nodos_explorados += 1   # Se incrementa el contador de nodos analizados

    score = evaluar_estado(tablero)

    # Condiciones de parada del algoritmo
    if abs(score) == 10 or es_tablero_lleno(tablero) or profundidad == PROFUNDIDAD_MAX:
        return score

    # CASO DEL JUGADOR MAX (X)
    if es_max:

        mejor = -math.inf

        for i in range(3):
            for j in range(3):

                if tablero[i][j] == " ":

                    tablero[i][j] = "X"

                    valor = alphabeta(tablero, profundidad+1, alfa, beta, False)

                    tablero[i][j] = " "

                    mejor = max(mejor, valor)

                    # Aplicación de la poda Alpha-Beta
                    if USAR_PODA:
                        alfa = max(alfa, mejor)
                        if beta <= alfa:
                            break

        return mejor


    # CASO DEL JUGADOR MIN (O)
    else:

        mejor = math.inf

        for i in range(3):
            for j in range(3):

                if tablero[i][j] == " ":

                    tablero[i][j] = "O"

                    valor = alphabeta(tablero, profundidad+1, alfa, beta, True)

                    tablero[i][j] = " "

                    mejor = min(mejor, valor)

                    # Aplicación de la poda Alpha-Beta
                    if USAR_PODA:
                        beta = min(beta, mejor)
                        if beta <= alfa:
                            break

        return mejor


# FUNCIÓN: mejor_movimiento
# Determina el mejor movimiento posible para la IA
# utilizando el algoritmo Minimax / Alpha-Beta
def mejor_movimiento(tablero):

    mejor_valor = -math.inf
    mejor_mov = None

    for i in range(3):
        for j in range(3):

            if tablero[i][j] == " ":

                tablero[i][j] = "X"

                valor = alphabeta(tablero, 0, -math.inf, math.inf, False)

                tablero[i][j] = " "

                if valor > mejor_valor:
                    mejor_valor = valor
                    mejor_mov = (i, j)

    return mejor_mov


# INICIALIZACIÓN DEL TABLERO
tablero = [
    [" "," "," "],
    [" "," "," "],
    [" "," "," "]
]

turno = "X"


# INICIO DE MEDICIÓN DE TIEMPO
inicio = time.time()


# BUCLE PRINCIPAL DEL JUEGO
while evaluar_estado(tablero) == 0 and not es_tablero_lleno(tablero):

    imprimir_tablero(tablero)

    # Turno del jugador humano
    if turno == "X":

        fila,col = map(int,input("fila col: ").split())
        tablero[fila][col] = "X"

    # Turno de la IA
    else:

        mov = mejor_movimiento(tablero)
        tablero[mov[0]][mov[1]] = "O"

    # Cambio de turno
    turno = "O" if turno=="X" else "X"


# FIN DEL JUEGO
fin = time.time()

imprimir_tablero(tablero)


# RESULTADOS DEL EXPERIMENTO
print("Tiempo ejecución:", fin-inicio)
print("Nodos explorados:", nodos_explorados)
print("Usando poda:", USAR_PODA)
print("Profundidad máxima:", PROFUNDIDAD_MAX)
```

```python

# IMPORTACIÓN DE LIBRERÍAS
import math
import time

# Guardamos el tiempo de inicio del programa
inicio = time.time()


# FUNCIÓN: imprimir_tablero
# Muestra el estado actual del tablero en la consola.
# Cada fila se imprime con espacios entre las casillas.
def imprimir_tablero(tablero):
    for fila in tablero:
        print(" ".join(fila))
    print()  # línea en blanco para separar visualmente


# FUNCIÓN: es_tablero_lleno
# Verifica si el tablero ya no tiene espacios vacíos.
# Retorna True si está lleno y False si aún hay espacios.
def es_tablero_lleno(tablero):
    for fila in tablero:
        if " " in fila:   # si existe una casilla vacía
            return False
    return True


# FUNCIÓN: evaluar_estado
# Evalúa el estado actual del juego de tic-tac-toe.
#
# Retorna:
#   1   -> gana X
#  -1   -> gana O
#   0   -> empate
# None  -> el juego continúa
def evaluar_estado(tablero):

    ganador = None

    # Revisar filas
    for fila in tablero:
        if fila.count("X") == 3:
            ganador = "X"
        elif fila.count("O") == 3:
            ganador = "O"

    # Revisar columnas
    for col in range(3):
        if tablero[0][col] == tablero[1][col] == tablero[2][col] and tablero[0][col] != " ":
            ganador = tablero[0][col]

    # Revisar diagonal principal
    if tablero[0][0] == tablero[1][1] == tablero[2][2] and tablero[0][0] != " ":
        ganador = tablero[0][0]

    # Revisar diagonal secundaria
    if tablero[0][2] == tablero[1][1] == tablero[2][0] and tablero[0][2] != " ":
        ganador = tablero[0][2]

    # Determinar el resultado
    if ganador == "X":
        return 1
    elif ganador == "O":
        return -1
    elif es_tablero_lleno(tablero):
        return 0
    else:
        return None


# CREACIÓN DEL TABLERO INICIAL
# Se representa como una matriz 3x3 con espacios vacíos
tablero_inicial = [
    [" ", " ", " "],
    [" ", " ", " "],
    [" ", " ", " "]
]

# El jugador X inicia el juego
turno = "X"


# BUCLE PRINCIPAL DEL JUEGO
# Se ejecuta mientras no haya ganador y el tablero
# no esté lleno
while evaluar_estado(tablero_inicial) is None and not es_tablero_lleno(tablero_inicial):

    # Mostrar el tablero actual
    imprimir_tablero(tablero_inicial)

    # Turno del jugador X
    if turno == "X":

        fila, col = map(int, input("Turno de X. Ingrese fila y columna (separadas por espacio): ").split())

        if tablero_inicial[fila][col] == " ":
            tablero_inicial[fila][col] = "X"
        else:
            print("Casilla ocupada. Intenta de nuevo.")
            continue

    # Turno del jugador O
    else:

        fila, col = map(int, input("Turno de O. Ingrese fila y columna (separadas por espacio): ").split())

        if tablero_inicial[fila][col] == " ":
            tablero_inicial[fila][col] = "O"
        else:
            print("Casilla ocupada. Intenta de nuevo.")
            continue

    # Cambio de turno entre jugadores
    turno = "O" if turno == "X" else "X"


# FIN DEL JUEGO
# Se imprime el tablero final y el resultado
imprimir_tablero(tablero_inicial)

resultado = evaluar_estado(tablero_inicial)

if resultado == 1:
    print("¡X gana!")
elif resultado == -1:
    print("¡O gana!")
else:
    print("¡Empate!")


# TIEMPO DE EJECUCIÓN
# Calcula cuánto tiempo tardó en ejecutarse el programa
fin = time.time()
print("Tiempo:", fin - inicio)
```
